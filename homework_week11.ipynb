```python
vec = CountVectorizer(analyzer = 'text', vocabulary = emolex_df.text,
                      lowercase=False,
                      strip_accents = 'unicode',
                      stop_words= list(nlp.Defaults.stop_words),
                      ngram_range=(1, 2))
```
